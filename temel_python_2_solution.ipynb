### 🔸 Soru 10:
`while` döngüsü ile basamak sayısı bulma: Verilen bir sayının kaç basamaklı olduğunu, sayıyı 10'a bölerek bulun (string dönüşümü kullanmadan).

```python
# Cevabınızı buraya yazın:
# Örnek: 4752 -> 4 basamaklı
sayi = 4752

basamak_sayisi = 0
while sayi > 0:
    sayi //= 10
    basamak_sayisi += 1

print(basamak_sayisi)
```

### 🔸 Soru 12:
Bir öğrenci listesi var. Her öğrenci `(isim, not)` tuple'ı olarak saklanıyor.
- Notu 70'in üzerinde olanları filtreleyin
- En yüksek notu alan öğrenciyi bulun
- Notların ortalamasını hesaplayın

```python
# Cevabınızı buraya yazın:
ogrenciler = [
    ("Ahmet", 85),
    ("Büşra", 62),
    ("Can", 91),
    ("Deniz", 74),
    ("Elif", 58),
    ("Furkan", 88)
]

basarili_listesi = [ogrenci for ogrenci in ogrenciler if ogrenci[1] > 70]
print("70 üzeri öğrenciler:", basarili_listesi)

en_yuksek_not_ogrenci = max(ogrenciler, key=lambda x: x[1])
print("En yüksek not alan öğrenci:", en_yuksek_not_ogrenci)

ortalama = sum([ogrenci[1] for ogrenci in ogrenciler]) / len(ogrenciler)
print("Notların genel ortalaması:", ortalama)
```

### 🔸 Soru 15:
Özyinelemeli (recursive) bir fonksiyon yazın: `faktoriyel(n)` - n! hesaplasın.
- 0! = 1
- n! = n × (n-1)!

5!, 10! değerlerini hesaplayın.

```python
# Cevabınızı buraya yazın:
def faktoriyel(n):
    if n == 0:
      return 1
    return n * faktoriyel(n - 1)

print(faktoriyel(3))
print(faktoriyel(5))
```

### 🔸 Soru 18 (Bütünleştirici):
Basit bir **öğrenci not sistemi** yazın:

- `ogrenciler` adında boş bir sözlük oluşturun (anahtar: isim, değer: notlar listesi)
- `ogrenci_ekle(isim)`: Yeni öğrenci ekler
- `not_ekle(isim, not_)`: Öğrenciye not ekler. Öğrenci yoksa hata mesajı verir.
- `ortalama_hesapla(isim)`: Öğrencinin not ortalamasını döndürür
- `sinif_raporu()`: Tüm öğrencileri ve ortalamalarını yazdırır, sınıf ortalamasını gösterir

En az 3 öğrenci ve her birine 3 not girerek test edin.

```python
# Cevabınızı buraya yazın:
ogrenciler = {}

def ogrenci_ekle(isim):
    if isim in ogrenciler:
        print(f"Uyarı: {isim} zaten sistemde kayıtlı bir öğrenci!")
    else:
        ogrenciler[isim] = []
        print(f"{isim} sisteme öğrenci olarak eklendi.")

def not_ekle(isim, not_):
    if isim not in ogrenciler:
        print(f"Hata: {isim} adlı öğrenci bulunamadı!")
        return

    if not_ < 0 or not_ > 100:
        print(f"Hata: Not 0-100 arasında olmalıdır! Girilen not: {not_}")
        return

    ogrenciler[isim].append(not_)
    print(f"{isim} adlı öğrenciye {not_} notu eklendi.")

def ortalama_hesapla(isim):
    if isim not in ogrenciler:
        print(f"Hata: {isim} adlı öğrenci bulunamadı!")
        return None

    notlar = ogrenciler[isim]

    # Eğer öğrencinin hiç notu yoksa
    if len(notlar) == 0:
        print(f"{isim} adlı öğrencinin henüz notu yok!")
        return 0

    ortalama = sum(notlar) / len(notlar)
    return round(ortalama, 2)

def sinif_raporu():
    if len(ogrenciler) == 0:
        print("Sistemde henüz hiç öğrenci yok!")
        return

    print("\n" + "="*60)
    print("SINIF RAPORU")
    print("="*60)

    ogrenci_sayisi = 0
    notlar_toplam = 0

    for isim, notlar in ogrenciler.items():
        ortalama = ortalama_hesapla(isim)
        if ortalama is not None:
            print(f"{isim:<20} | Notlar: {notlar} | Ortalama: {ortalama}")
            notlar_toplam += ortalama
            ogrenci_sayisi += 1

    if ogrenci_sayisi > 0:
        sinif_ortalamasi = notlar_toplam / ogrenci_sayisi
        print("-"*60)
        print(f"Sınıf Ortalaması: {round(sinif_ortalamasi, 2)}")
    print("="*60 + "\n")


# Test kodu:
ogrenci_ekle("Ahmet")
ogrenci_ekle("Büşra")
ogrenci_ekle("Can")
not_ekle("Ahmet", 85)
not_ekle("Ahmet", 90)
not_ekle("Ahmet", 78)
not_ekle("Büşra", 92)
not_ekle("Büşra", 88)
not_ekle("Büşra", 95)
not_ekle("Can", 70)
not_ekle("Can", 65)
not_ekle("Can", 73)
sinif_raporu()
```
