### 🔸 Soru 1:
Aşağıdaki değişkenleri tanımlayın ve her birinin tipini `type()` fonksiyonu ile yazdırın:

## 3️⃣ String (Metin) İşlemleri

```python
# Örnek: String metodları
metin = "Merhaba Python!"

print(metin.upper())       # Büyük harf
print(metin.lower())       # Küçük harf
print(metin.replace("Python", "Dünya"))  # Değiştirme
print(len(metin))          # Uzunluk
print(metin[0:7])          # Dilimleme
```

### 🔸 Soru 5:
`cumle = "  python programlama dili çok güçlüdür  "` değişkeni için:
- Baştaki ve sondaki boşlukları temizleyin (`.strip()`)
- İlk harfini büyük yapın (`.capitalize()`)
- Kaç kelimeden oluştuğunu bulun (`.split()` ve `len()` kullanın)
- "güçlü" kelimesinin cümlede olup olmadığını kontrol edin (`in` operatörü)

```python
# Cevabınızı buraya yazın:
cumle = "  python programlama dili çok güçlüdür  "
cleaned_sentence = cumle.strip()
print(cleaned_sentence)
capitalized_sentence = cleaned_sentence.capitalize()
print(capitalized_sentence)
print(len(capitalized_sentence.split()))
print("güçlü" in capitalized_sentence)
```

### 🔸 Soru 8:
Bir sayının 2'ye ve 3'e tam bölünüp bölünmediğini kontrol edin. 4 farklı durum vardır:
- Hem 2'ye hem 3'e bölünebilir
- Sadece 2'ye bölünebilir
- Sadece 3'e bölünebilir
- Hiçbirine bölünemiyor

```python
# Cevabınızı buraya yazın:
n = 12

if n % 2 == 0 and n % 3 == 0:
    print("Sayı, Hem 2'ye hem 3'e bölünebilir")
elif n % 2 == 0:
    print("Sayı, Sadece 2'ye bölünebilir")
elif n % 3 == 0:
    print("Sayı, Sadece 3'e bölünebilir")
else:
    print("Sayı, Hiçbirine bölünemiyor")
```

### 🔸 Soru 10:
`while` döngüsü ile kullanıcının girdiği sayıya (n=5) kadar Fibonacci dizisini yazdırın.

Örnek çıktı: `0, 1, 1, 2, 3`

```python
# Cevabınızı buraya yazın:
n = 5

a, b = 0, 1
print(a, end=", ")
while b < n:
    print(b, end=", ")
    a, b = b, a+b
```

### 🔸 Soru 11 (İç İçe Döngü):
5x5 çarpım tablosunu yazdırın. Her satır şu formatta olsun:

`1x1=1  1x2=2  1x3=3 ...`

```python
# Cevabınızı buraya yazın:

for i in range(1,6):
  for j in range(1,6):
    print(f"{i}x{j}={i*j}", end="\t")
  print()
```

### 🔸 Soru 13 (List Comprehension):
List comprehension kullanarak:
- 1-20 arasındaki çift sayıların listesini oluşturun
- Bir kelime listesindeki 4'ten uzun kelimeleri filtreleyin

```python
# Cevabınızı buraya yazın:
kelimeler = ["at", "kaplan", "it", "fil", "kedi", "aslan", "köpek"]

for i in range(1, 21):
  if i % 2 == 0:
    print(i, end=" ")
print()

for kelime in kelimeler:
  if len(kelime) > 4:
    print(kelime, end=" ")
```

### 🔸 Soru 15:
Bir sayının asal olup olmadığını kontrol eden `asal_mi(n)` fonksiyonu yazın.
- 1'den 20'ye kadar olan asal sayıları bu fonksiyonu kullanarak yazdırın.

```python
# Cevabınızı buraya yazın:
def asal_mi(n):
  for i in range(2, n):
    if n % i == 0:
      return False
  return True

for i in range(1, 21):
  if i == 1:
    continue
  if asal_mi(i):
    print(i, end=" ")
    pass  # buraya kodunuzu yazın
```

### 🔸 Soru 16:
Bir listenin ortalamasını, en büyük ve en küçük elemanını döndüren `liste_istatistik(liste)` fonksiyonu yazın.
Fonksiyon 3 değer döndürsün: `(ortalama, minimum, maksimum)`

```python
# Cevabınızı buraya yazın:
def liste_istatistik(liste):

  ortalama = sum(liste) / len(liste)
  minimum = min(liste)
  maksimum = max(liste)
  return ortalama, minimum, maksimum

test = [4, 8, 15, 16, 23, 42]
ort, mini, maksi = liste_istatistik(test)
print(f"Ortalama: {ort}, Min: {mini}, Max: {maksi}")
```

### 🔸 Soru 17:
Kullanıcıdan bir sayı alındığını varsayın (`girdi = "abc"`).
Bu değeri int'e çevirmeye çalışırken `ValueError` hatası oluşabilir.
try-except ile bunu yakalayan ve anlamlı bir mesaj veren kod yazın.

```python
# Cevabınızı buraya yazın:
girdi = "abc"

try:
  int(girdi)
except ValueError:
  print("Girdi bir sayı değil, lütfen sayı girdiğinizden emin olun.")
```

### 🔸 Soru 18 (Bütünleştirici):
Bir **not defteri uygulaması** yazın:
- `notlar` adında boş bir liste oluşturun
- `not_ekle(metin)` fonksiyonu: Notu listeye ekler
- `not_listele()` fonksiyonu: Tüm notları numaralandırarak yazdırır
- `not_sil(index)` fonksiyonu: Verilen indexteki notu siler, hatalı index için hata mesajı verir

Fonksiyonları test edin: 3 not ekleyin, listeleyin, ortadakini silin, tekrar listeleyin.

```python
# Cevabınızı buraya yazın:
notlar = []

def not_ekle(metin):
  notlar.append(metin)
  pass

def not_listele():
  for i in notlar:
    print(i)

def not_sil(index):
  del notlar[index]
  pass

# Test kodu:
not_ekle("Python öğren")
not_ekle("Alıştırmaları tamamla")
not_ekle("Proje yap")
not_listele()
not_sil(1)
not_listele()
```
